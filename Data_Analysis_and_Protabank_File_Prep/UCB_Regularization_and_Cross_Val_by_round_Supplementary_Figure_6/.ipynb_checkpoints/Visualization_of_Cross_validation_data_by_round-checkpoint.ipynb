```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import pickle
```

```python
#Load all the data 

lam_array,cc_array,err2_array,lam,cc,err2,Yact,Y_hat_cv=pickle.load(open('r2_regularization_and_xval_data.p','rb'))
```

```python
#Build the figure. Loop through the data files and load each round individually

plt.figure(figsize=(15,45))

nrows=10
ncols=3

for i in range(2,11):

    #Load the data from the given round
    lam_array,cc_array,err2_array,lam,cc,err2,Yact,Y_hat_cv=pickle.load(open('r'+str(i)+'_regularization_and_xval_data.p','rb'))
    
    #Make the plots
    #Lambda scan (CC)
    plt.subplot(nrows,ncols,ncols*i-ncols-2)
    plt.plot(lam_array,cc_array,'-',color='0.7',label='CC')
    plt.plot(lam,cc,'o',label='Selected Lambda (CC)',color='k')
    plt.ylabel('Correlation Coefficient')
    #plt.xlabel('Lambda')
    
    if i>=4:
        left,right=plt.xlim()
        plt.xlim((-.0025,.1))
        
    if i==10:
        plt.xlabel('Lambda')
    
    #Lambda scan (Squared error)
    plt.subplot(nrows,ncols,ncols*i-ncols-1)
    plt.title('Round ' + str(i))

    plt.plot(lam_array,err2_array,'-',color='0.7',label='Squared Error')
    plt.plot(lam,err2,'o',label='Selected Lambda (Squared Error)',color='k')
    plt.ylabel('Sum Squared Error')
    #plt.xlabel('Lambda')

    
    if i>=4:
        left,right=plt.xlim()
        plt.xlim((-.0025,.1))
        
    if i==10:
        plt.xlabel('Lambda')
    
    #LOOCV
    plt.subplot(nrows,ncols,ncols*i-ncols)
    sns.scatterplot(Yact[:,0],Y_hat_cv[:,0],marker='o',color='0.5',edgecolor='k')
    plt.plot([np.min(Yact),np.max(Yact)],[np.min(Yact),np.max(Yact)],'k--')
    #plt.xlabel('Observed Activity')
    plt.ylabel('Predicted Activity')
    
    if i==10:
        plt.xlabel('Observed Activity')
    
        
```
